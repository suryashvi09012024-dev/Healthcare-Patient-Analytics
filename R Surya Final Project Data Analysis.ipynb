# **Stage 3 - EDA and Visulalization**

# **Univariate Analysis**

```python
# Count Plot
plt.figure(figsize=(6,4))
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

# Interpretation - Gender Distribution
 The count plot represents the number of patients in each gender category. It helps identify which gender has more patients and provides a clear comparison of the gender distribution in the dataset.

```python
# Histogram Plot
plt.figure(figsize=(8,5))
plt.hist(df['treatment_cost'], bins=15, edgecolor='black')
plt.title('Treatment Cost Distribution')
plt.xlabel('Treatment Cost')
plt.ylabel('Frequency')
plt.show()
```

# Interpretation - Treatment Cost Distribution
 The histogram illustrates the frequency distribution of treatment costs. It helps identify the range where most treatment costs are concentrated and shows whether the data is evenly distributed or skewed.

```python
# Box Plot
plt.figure(figsize=(8,5))
sns.boxplot(y='treatment_cost', data=df)
plt.title('Treatment Cost Box Plot')
plt.show()
```

# Interpretation - Treatment Cost
 The box plot represents the distribution of treatment costs. The line inside the box indicates the median, the box shows the middle 50% of the data, and the points outside the box represent outliers. It helps identify the spread and unusual values in treatment costs.

# **Bivariate Analysis**

```python
# Bar Plot
plt.figure(figsize=(6,4))
sns.countplot(x='age_group', data=df)
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Patients')
plt.show()
```

# Interpretation- Age Group Distribution
 The Age Group Distribution chart shows the number of patients in each age group. It helps identify which age group has the highest and lowest patient count. This analysis provides an overview of the patient population and supports understanding of the age-wise distribution."

```python
# Scatter Plot
plt.figure(figsize=(8,5))
sns.scatterplot(x='treatment_cost', y='recovery_score', data=df)
plt.title('Treatment Cost vs Recovery Score')
plt.xlabel('Treatment Cost')
plt.ylabel('Recovery Score')
plt.show()
```

# Interpretation - Treatment Cost vs Recovery Score
 The scatter plot compares treatment cost with recovery score. It shows how these two variables are related and helps identify whether there is a positive, negative, or weak relationship between them.

#

```python
# Correlation Heatmap
df['length_of_stay_days'] = pd.to_numeric(
    df['length_of_stay_days'],
    errors='coerce'
)

plt.figure(figsize=(8,6))
sns.heatmap(
    df[['length_of_stay_days',
        'treatment_cost',
        'recovery_score']].corr(),
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Correlation Heatmap')
plt.show()

```

# Interpretation – Correlation Heatmap

The correlation heatmap shows the relationship between the numerical variables in the healthcare dataset. Treatment cost, recovery score, and length of stay days are compared using correlation values. Positive correlation values indicate that two variables increase together, while negative correlation values indicate an inverse relationship.

# **Multivariate Analysis**

```python
# Pair Plot
sns.pairplot(
    df[['length_of_stay_days',
        'treatment_cost',
        'recovery_score',
        'gender']],
    hue='gender'
)

plt.show()
```

# Interpretation (Report)

The pair plot shows the relationship between the numeric variables. The diagonal plots show the distribution of each variable, while the scatter plots show how pairs of variables are related. It helps identify trends, clusters, and possible correlations among the variables.

```python
# Pie Plot
plt.figure(figsize=(6,6))
df['age_group'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Age Group Distribution')
plt.ylabel('')
plt.show()
```

# Interpretation:

The pie chart shows the percentage distribution of patients across different age groups. It helps identify which age group represents the largest and smallest share of patients.

```python
#Bar Chart (Department-wise Patient Count)
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.countplot(x='department', data=df, palette='viridis')

plt.title('Patients by Department')
plt.xlabel('Department')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45)

plt.show()
```

# Interpretation

The bar chart shows the number of patients in each hospital department. It helps identify which department receives the highest and lowest patient visits. This analysis supports understanding patient demand across different departments and can help in resource planning.

```python
#Violin Plot (Treatment Cost by Age Group)
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.violinplot(x='age_group',
               y='treatment_cost',
               data=df,
               palette='Set2')

plt.title('Treatment Cost by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Treatment Cost')

plt.show()
```

# Interpretation

The violin plot shows the distribution of treatment costs across different age groups. It highlights the spread, density, and variation in treatment costs for each age group. This helps identify whether certain age groups tend to have higher or more variable treatment expenses.
